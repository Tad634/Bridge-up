# Final project for Climate Change

```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
csvfile = open("CloudFraction1_2018.csv", "r")
csvReader = csv.reader(csvfile, delimiter=',')
data_lists = list(csvReader)
```

```python
width = len(data_lists[0])
height = len(data_lists)
grid = np.empty([height, width, 3], dtype=np.uint8) 
```

```python
all_data = []
for r, row in enumerate(data_lists):
    for c, value in enumerate(row):
        if float(value) < 99999:
            all_data.append(float(value))
        if float(value) <= 0:
            grid[r, c] = [22, 22, 130]
        elif float(value) <= 0.2:
            grid[r, c] = [100, 100, 207]
        elif float(value) <= 0.4:
            grid[r, c] = [151, 151, 217]
        elif float(value) <= 0.6:
            grid[r, c] = [161, 191, 235]
        elif float(value) <= 0.8:
            grid[r, c] = [214, 214, 255]
        elif float(value) <= 1 :
            grid[r, c] = [255, 255, 255]
        else:
            grid[r, c] = [0, 0, 0]
```

```python
print(min(all_data))
print(max(all_data))
```

```python
plt.scatter([],[], c = (22/255, 22/255, 130/255), label = "0%")
plt.scatter([],[], c = (100/255, 100/255, 207/255), label = "20%")
plt.scatter([],[], c = (151/255, 151/255, 210/255), label = "40%")
plt.scatter([],[], c = (161/255, 191/255, 235/255), label = "60%")
plt.scatter([],[], c = (214/255, 214/255, 255/255), label = "80%")
plt.scatter([],[], color = "white", edgecolor="black", label="100%")
plt.legend(bbox_to_anchor = [1.21, 1.0])
plt.title("Cloud Cover (2018)")
plt.imshow(grid)
plt.axis('off')
plt.savefig("Cloud_Fraction_2018.png",dpi = 1000)
plt.show()
```
